### **=** is   Assignment Operator 

```python
a = 8 # the value 8 gets assigned to a
```

```python
print('a =',a)
print('type of a =', type(a))
```

```python
a = 9 # a is overwritten by 9
print('so now a = ',a)
```

```python
a = 'hello' # a is overwritten by 9
print('so now a = ',a)
print('type of a =', type(a))
```
